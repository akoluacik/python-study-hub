```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name" : ["John", "Gabriel", "Nicolo", "Antonio", "Wendel", "Goytacaz"],
    "Activity" : ["Jogging", "Swimming", "Jogging", "Basketball", "Football", "Football"],
    "Calorie" : [100, 200, 300, 400, 500, 600]
}

df1 = pd.DataFrame(data)
df1
```

```python
data2 = {
    "Name" : ["Roques", "Gerrard", "Nathan", "Olivier", "Goncalo", "Abdelhak"],
    "Activity" : ["Jogging", "Swimming", "Jogging", "Basketball", "Football", "Football"],
    "Calorie" : [170, 290, 350, 450, 550, 600]
}

df2 = pd.DataFrame(data2, index = np.arange(6, 12))
df2
```

```python
data3 = {
    "Name" : ["Olivier", "Dominic", "Jean"],
    "Activity" : ["Jogging", "Swimming", "Jogging"],
    "Calorie" : [100, 200, 300]
}

df3 = pd.DataFrame(data3, index=np.arange(12,15))
df3
```

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df2, df3], axis=1)
```
