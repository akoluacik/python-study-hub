#### Assume there are two tables, whose some columns have the same name and same data, while some columns are different. If you merge these tables, you have a table with all columns.

```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name" : ["John", "Gabriel", "Nicolo", "Antonio", "Wendel", "Goytacaz"],
    "Activity" : ["Jogging", "Swimming", "Jogging", "Basketball", "Football", "Football"],
    "Calorie" : [100, 200, 300, 400, 500, 600]
}

df1 = pd.DataFrame(data)
df1
```

```python
data2 = {
    "Name" : ["John", "Gabriel", "Nicolo", "Antonio", "Wendel", "Goytacaz"],
    "Activity" : ["Jogging", "Swimming", "Jogging", "Basketball", "Football", "Football"],
    "Wage" : [150, 200, 305, 486, 573, 601]
}

df2 = pd.DataFrame(data2)
df2
```

```python
pd.merge(df1, df2, on=["Name", "Activity"])
```
