```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(-100, 100, 200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])

ax.plot(x, 2*x ,label="y=2*x")
ax.plot(x, x**2, label="y=x^2")
ax.legend(loc="best")
```

```python
fig = plt.figure(figsize=(3,3))

ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])

ax.plot(x, 2*x ,label="y=2*x")
ax.plot(x, x**2, label="y=x^2")
ax.legend(loc="best")
```

```python
# dpi determines the quality of the graph
# it is good to determine the value when you finish your job and want to
# save it.
fig = plt.figure(figsize=(3,3), dpi=120)

ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])

ax.plot(x, 2*x ,label="y=2*x")
ax.plot(x, x**2, label="y=x^2")
ax.legend(loc="best")
```

```python
fig.savefig("myFig2.png", dpi=200)
```

```python
fig = plt.figure(figsize=(3,3), dpi=120)

ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])

ax.plot(x, 2*x ,label="y=2*x", color="#000")
ax.plot(x, x**2, label="y=x^2", color="#00FFFF")
```

```python
fig = plt.figure(figsize=(3,3), dpi=120)

ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])

#alpha = transperancy
ax.plot(x, 2*x ,label="y=2*x", color="#000", alpha = 0.3)
ax.plot(x, x**2, label="y=x^2", color="#00FFFF", alpha = 0.9)
```

```python
fig = plt.figure(figsize=(3,3), dpi=120)

ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])

ax.plot(x, 2+x ,label="y=2*x", color="#000", linewidth=2.0)
ax.plot(x, x*2+3, label="y=x*2+3", color="#00FFFF",linewidth=3.5)
```

```python
(fig,ax) = plt.subplots()

ax.plot(x, 2+x ,label="y=2*x", color="#000", linewidth=2.0, linestyle='--')
ax.plot(x, np.tan(x)+3, label="y=x**2+3", color="#00FFFF",linewidth=3.5, linestyle='-')
ax.plot(x, x*5+4, label="y=x*2+3", color="#23670F",linewidth=3.5, linestyle='-.')
ax.plot(x, -x*7+5, label="y=x*2+3", color="#636422",linewidth=3.5, linestyle=':')
```

```python
(fig,ax) = plt.subplots()

ax.plot(x, 2+x ,label="y=2*x", color="#000", linewidth=2.0, linestyle='--')
ax.plot(x, np.tan(x)+3, label="y=tan(x)+3", color="#00FFFF",linewidth=3.5, linestyle='-')
ax.plot(x, x*5+4, label="y=x*2+3", color="#23670F",linewidth=3.5, linestyle='-.')
ax.plot(x, -x*7+5, label="y=x*2+3", color="#636422",linewidth=3.5, linestyle=':', marker="o", markerfacecolor="red", markersize=4)
ax.legend(loc="best")
```
