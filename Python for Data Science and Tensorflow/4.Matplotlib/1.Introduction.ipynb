```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(1, 100, 50)
y = np.logspace(1, 100, 50)
y
```

```python
plt.plot(x, y, 'b*-')
```

#### Creating axes

```python
myFigure = plt.figure()
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2, 0.2, 0.9, 0.9])
figureAxes.plot(x, y, "g")
figureAxes.set_xlabel("X axis")
figureAxes.set_ylabel("Y axis")
figureAxes.set_title("Graphic Title")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.5, 0.5, 0.9, 0.9])
figureAxes.plot(x, y, "g")
figureAxes.set_xlabel("X axis")
figureAxes.set_ylabel("Y axis")
figureAxes.set_title("Graphic Title")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2, 0.2, 0.3, 0.3])
figureAxes.plot(x, y, "g")
figureAxes.set_xlabel("X axis")
figureAxes.set_ylabel("Y axis")
figureAxes.set_title("Graphic Title")
```

#### To understand the argument of the add_axes method

```python
figure = plt.figure()
ax1 = figure.add_axes([0.0, 0.0, 0.8, 0.7])
ax2 = figure.add_axes([0.2, 0.2, 0.3, 0.3])
#ax2 = figure.add_axes([0.0, 0.0, 0.3, 0.3])

ax1.plot(x,y,"g")
ax1.set_xlabel("X Axis")
ax1.set_ylabel("Y Axis")
ax1.set_title("Outer Graph Title")

ax2.plot(y,x,"r")
ax2.set_xlabel("X Axis")
ax2.set_ylabel("Y Axis")
ax2.set_title("Inner Graph Title")

```
