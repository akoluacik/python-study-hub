**Q7 (Duplicate Elimination)** Create a function that receives a list and returns a (possibly shorter) list containing only the unique values in sorted order. Test your function with a list of numbers and a list of strings.

````python
def duplicate(lst):
    res = []
    for element in lst:
        if element not in res:
            res.append(element)
        
    return sorted(res)
    
a = [2,5,8,4,5,6,8,2,3,4,2,3]
res = duplicate(a)
res
````

**Q9 (Palindrome Tester)** A string that’s spelled identically backward and forward, like 'radar', is a palindrome. Write a function is_palindrome that takes a string and returns True if it’s a palindrome and False otherwise. Use a stack (simulated with a list as we did in Section 5.11) to help determine whether a string is a palindrome. Your function should ignore case sensitivity (that is, 'a' and 'A' are the same), spaces and punctuation. 

````python
def is_palindrome(string:str): # very easy way
    return string == string[::-1] 

def is_palindrome_with_stack(string:str):
    str_lowered = string.lower()
    
    stack=list(str_lowered[::-1]) # hold reverse of the string as a list
    str_list = list(str_lowered) # hold the string as a list
    
    while len(stack):
        x = stack.pop()    # pop the elements
        y = str_list.pop() # pop the elements
        if x != y:         # if elements are not the same return false
            return False
    return True
is_palindrome_with_stack("Efe")
````

````python
def summarize_letters(string:str):
    res = [("",0)] * 26
    for letter in string:
        res[ord(letter) - ord('a')] = list(res[ord(letter) - ord('a')])
    
        if res[ord(letter) - ord('a')][0] == "":
                res[ord(letter) - ord('a')][0] += letter
                res[ord(letter) - ord('a')][1] = 1 
            
        else:
            res[ord(letter) - ord('a')][1] += 1
    
        res[ord(letter) - ord('a')] = tuple(res[ord(letter) - ord('a')])
    return res

res = summarize_letters("abcdefghijklmnoprsttuvyzqxw")

"""Checks if contains all the alphabets"""
contains = True
for x in res:
    if all(x) == False:
        control = False
        print("Does not contain all the alphabets")
        break

if contains:
    print("Contains all the alphabets")
            
````

**Q14 (Is a Sequence Sorted?):** Create a function is_ordered that receives a sequence and returns True if the elements are in sorted order. Test your function with sorted and unsorted lists, tuples and strings. 

````python
def is_ordered(sequence):
    if sequence is not list:
        sequence = list(sequence)
    return sorted(sequence) == sequence

print(is_ordered([1,3,2]))
print(is_ordered((1,3,2)))
print(is_ordered('132'))
print(is_ordered([1,2,4]))
print(is_ordered((1,2,3)))
print(is_ordered('258'))
````

**Q17(Filter/Map Performance)**: With regard to the following code:

```python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
list(map(lambda x: x ** 2, 
filter(lambda x: x % 2 != 0, numbers)))
```

a) How many times does the filter operation call its lambda argument?
b) How many times does the map operation call its lambda argument?
c) If you reverse the filter and map operations, how many times does the map operation call its lambda argument?

We can use *print* and *user-defined* functions to show the moments of the functions passed to map filter being called. Here we have **global i** and **global j** variables demonstrates how many times is the corresponding function invoked. Our *user-defined* functions are operating the same way as the ones written with *lambda* expression are.

````python
i=0
j=0
def foo(x):
    global i
    i += 1
    print(f'function foo is called {i}th times')
    return x%2 != 0


def bar(x):
    global j
    j += 1
    print(f'function bar is called {j}th times')
    return x**2

numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
list(map(bar,  filter(foo, numbers)))
````

**a) and b)**
So *foo* function, which is the same as ```lambda x:x%2!=0```, is invoked **10 times** because it checks if the element is **odd** or not. *bar* function, the same as ```lambda x:x**2```, invoked 5 times because our array is filtered and reduced to *5-element* array.

**c)** This time, they are invoked *consecutively*.

````python
i=0
j=0
list(filter(bar,  map(foo, numbers)))
````

**Q22 (Simulating a Queue with a List)**: Simulate a queue of integers using list methods append (to simulate enqueue) and pop with the argument 0 (to simulate dequeue). Enqueue the values 3, 2 and 1, then dequeue them to show that they’re removed in FIFO order.

````python
import random

lst = []
print('Enqueue operation:')
for _ in range(20):
    rand_number = random.randint(1,100)
    lst.append(rand_number)
    print(f'list is now: {lst}')
print('Dequeue operation:')
for _ in range(20):
    lst.pop(0)
    print(f'list is now: {lst}')
````
